```python
# function to get genres in an array
def getGenres(text):
    data = json.loads(text)
    genres = []
    for i in range(len(data)):
        name = data[i].get('name')
        if(name == 'Action' or name == 'Adventure' or name == 'Comedy' or name == 'Crime' or name == 'Drama' or name == 'Romance' or name == 'Thriller'):
            genres.append(name)
    return genres
```

```python
# function to get actors in an array
def getActors(text):
    data = json.loads(text)
    actors = []
    for i in range(len(data)):
        name = data[i].get('name')
        actors.append(name)
    return actors
```

```python
# function to get production companies in an array
def getProduction_Comp(text):
    data = json.loads(text)
    prod_comp = []
    for i in range(len(data)):
        name = data[i].get('name')
        prod_comp.append(name)
    return prod_comp
```

```python
# function to get keywords in an array
def getKeywords(text):
    data = json.loads(text)
    keywords = []
    for i in range(len(data)):
        name = data[i].get('name')
        keywords.append(name)
    return keywords
```

```python
#convert the columns from dictionaries to arrays
df['genres_x']= df['genres_x'].apply(lambda x: getGenres(x))
df['cast_x']=df['cast_x'].apply(lambda x: getActors(x))
df['production_companies_x']=df['production_companies_x'].apply(lambda x: getProduction_Comp(x))
df['keywords_x']=df['keywords_x'].apply(lambda x: getKeywords(x))
```

```python
#import nlp libraries
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from string import punctuation
from nltk.tokenize.treebank import TreebankWordDetokenizer
from sklearn.feature_extraction.text import CountVectorizer

stop_words = set(stopwords.words('english'))
porter = PorterStemmer()
```

```python
#functions to clean text, remove stopwords, and get stem words
def cleanText(string):
    if type(string) == str:
        no_punct = ''.join([character for character in string if character not in punctuation])
        no_digit = ''.join([character for character in no_punct if not character.isdigit()]).lower()
        return no_digit
    else:
        return ""

def remove_stopwords(text):
    words = [word for word in text if not word in stop_words] 
    return words

def stemmer(text):
    stemmed_document = [porter.stem(word) for word in text]
    return stemmed_document
```

```python
#tokenize overview columns
clean_df['overview_x'] = clean_df['overview_x'].apply(lambda x: cleanText(x))

overview_tokenized = clean_df['overview_x'].apply(word_tokenize)
overview_no_stopwords = overview_tokenized.apply(remove_stopwords)
overview_stemmed = overview_no_stopwords.apply(stemmer)

overview_detokenized = overview_stemmed.apply(TreebankWordDetokenizer().detokenize)
```

```python
#tokenize keywords columns
keywords_detokenized1 = clean_df['keywords_x'].apply(TreebankWordDetokenizer().detokenize)

keywords_tokenized = keywords_detokenized1.apply(word_tokenize)
keywords_no_stopwords = keywords_tokenized.apply(remove_stopwords)
keywords_stemmed = keywords_no_stopwords.apply(stemmer)

keywords_detokenized = keywords_stemmed.apply(TreebankWordDetokenizer().detokenize)
```

```python
##tokenize production companies columns
production_companies_detokenized1 = clean_df['production_companies_x'].apply(TreebankWordDetokenizer().detokenize)
production_companies_tokenized = production_companies_detokenized1.apply(word_tokenize)
production_companies_no_stopwords = production_companies_tokenized.apply(remove_stopwords)
production_companies_stemmed = production_companies_no_stopwords.apply(stemmer)

production_companies_detokenized = production_companies_stemmed.apply(TreebankWordDetokenizer().detokenize)
```

```python
#tokenize tagline columns
clean_df['tagline_x'] = clean_df['tagline_x'].apply(lambda x: cleanText(x))
tagline_tokenized = clean_df['tagline_x'].apply(word_tokenize)
tagline_no_stopwords = tagline_tokenized.apply(remove_stopwords)
tagline_stemmed = tagline_no_stopwords.apply(stemmer)

tagline_detokenized = tagline_stemmed.apply(TreebankWordDetokenizer().detokenize)
```

```python
#tokenize cast columns
cast_detokenized = clean_df['cast_x'].apply(TreebankWordDetokenizer().detokenize)
```

```python
#OSR2 function
def OSR2(model, df_train, df_test, dependent_var):   
    y_test = df_test[dependent_var]
    y_pred = model.predict(df_test)
    SSE = np.sum((y_test - y_pred)**2)
    SST = np.sum((y_test - np.mean(df_train[dependent_var]))**2)    
    return 1 - SSE/SST
```
